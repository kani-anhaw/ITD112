### **Exercise 1.02: Indexing, Slicing, Splitting, and Iterating**

```python
import numpy as np

dataset1 = np.genfromtxt('normal_distribution_splittable.csv', delimiter=',')
dataset1
```

##### **Indexing**

```python
# indexing the 2nd row of the dataset
second_row = dataset1[1]
np.mean(second_row)
```

```python
# indexing the last element of the dataset 
last_row = dataset1[-1]
np.mean(last_row)
```

```python
# indexing the first value of the first row
first_val_first_row = dataset1[0][0]
np.mean(first_val_first_row)
```

```python
# indexing the last value of the second to the last 
last_val_second_last_row = dataset1[-2, -1]
np.mean(last_val_second_last_row)
```

##### **Slicing**

```python
#slicing an intersection of 4 elements (2x2) of the first two rows ad first twwo columns
subsection_2x2 = dataset1[1:3, 1:3]
np.mean(subsection_2x2)
```

```python
#selecting every second element of the fifth row
every_other_elem = dataset1[4, ::2]
np.mean(every_other_elem)
```

```python
# reversing the entry order, selecting the first two rows in reversed order
reversed_last_row = dataset1[-1, ::-1]
np.mean(reversed_last_row)
```

##### **Splitting**

```python
# splitting up our dataset horizontally on indices one third and two thirds 
hor_splits = np.hsplit(dataset1, (3))
```

```python
# splitting up our dataset vertically on index 2
ver_splits = np.vsplit(hor_splits[0], (2))
```

```python
#requested subsection of our dataset which has only half the amount of rows and only a third of the columns
print("Dataset", dataset1.shape)
print("Subset", ver_splits[0].shape)
```

##### **Iterating**

```python
#interating over whole dataset (eah value in each row)
curr_index = 0
for x in np.nditer(dataset1):
    print(x, curr_index)
    curr_index += 1
```

```python
# interating over whole dataset with indices matching the position in the dataset
for index, value in np.ndenumerate(dataset1):
    print(index, value)
```
